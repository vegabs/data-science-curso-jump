Antes de comenzar, vamos a hablar un poco de esta libreria o modulo.

**Numpy** es una librería optimizada para realizar cálculos numéricos con vectores y matrices. A diferencia de otros lenguajes de programación, Python no posee en su estructura central la figura de matrices. Eso quiere decir que para poder trabajar con esta estructura de datos deberiamos trabajar con listas de listas. NumPy introduce el concepto de arrays o matrices.

```python
from numpy import random

arreglo_5 =random.randint(1000, size=(100))
print(f"Media: {np.mean(arreglo_5)}")
print(f"Varianza: {np.var(arreglo_5)}")
```

```python
suma = np.sum(arreglo_5)
media = suma / arreglo_5.size
media
```

```python
# Cómo se calcula la varianza ------------------------------------------------
# Calcula la media de los datos.
# Encuentra la diferencia de cada punto de datos con respecto al valor medio.
# Eleva al cuadrado cada uno de estos valores.
# Suma todos los valores elevados al cuadrado.
# Divide esta suma de cuadrados entre n – 1 (para una muestra) o N (para la población).

sumatoria = arreglo_5 - np.mean(arreglo_5)
sumatoria = sumatoria ** 2
sumatoria_2 = np.sum(sumatoria)
varianza = sumatoria_2 / sumatoria.size
varianza
```
